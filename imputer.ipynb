```python
import numpy as np
```

```python
X_np = np.array([['PhD'],['12th'],['Btech'],['PhD']])
X_np
```

```python
import pandas as pd
X_df = pd.DataFrame(X_np,columns = ['Quals'])
X_df
```

```python
#ordinal encoding
from sklearn.preprocessing import OrdinalEncoder
OE = OrdinalEncoder()
X_OE = OE.fit_transform(X_df)
X_OE
```

```python
#explicitly mention the order of encoding
OE2 = OrdinalEncoder(categories=[['12th','Btech','PhD']])
X_OE2 = pd.DataFrame(OE2.fit_transform(X_df),columns = ['Quals'])
X_OE2
```

```python
X = np.array([['red'],['green'],['blue'],['red'],['blue']])
X
```

```python
X_df = pd.DataFrame(X,columns = ['colours'])
X_df
```

```python
#one hot encoder
from sklearn.preprocessing import OneHotEncoder
OHE = OneHotEncoder()
OHE_X = OHE.fit_transform(X_df)
OHE_X.toarray()
```

```python
OHE = OneHotEncoder(sparse = False)
OHE_X1 = OHE.fit_transform(X_df)
OHE_X1
OHE_X1 = pd.DataFrame(OHE_X1,columns = ['blue','green','red'])
OHE_X1
```

```python
cols = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','num']
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data",names = cols)
df
```

```python
df.ca.value_counts()
```

```python
df.replace('?',np.nan,inplace=True)
df
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'mean')
df = imputer.fit_transform(df)
df = pd.DataFrame(df,columns = cols)
df.isnull().sum()
```
